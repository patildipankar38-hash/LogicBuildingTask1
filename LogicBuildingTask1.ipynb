```python
username = "admin"
password = "1234"

# User input
username1 = input("Enter username: ")
password2 = input("Enter password: ")

# Login check
if username1 == username and password2 == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]

pass_student = 0
fail_student = 0

for i in marks:
    if i >= 50:
        pass_student += 1
    else:
        fail_student += 1

print("Total Students:", len(marks))
print("Number of Students Passed:", pass_student)
print("Number of Students Failed:", fail_student)
```

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)
print("Origional names:",names)
print("Cleaned Names List:", cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length > 10:
        print("Status: Longer than 10 characters")
```
